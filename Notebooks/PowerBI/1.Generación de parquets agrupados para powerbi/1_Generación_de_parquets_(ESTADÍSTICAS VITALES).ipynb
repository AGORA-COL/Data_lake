```python
# Importa pandas
import pandas as pd

# la ruta al archivo Parquet original
file_path = '/content/VITALESpart.0_2022.parquet'

# Carga el archivo Parquet en un DataFrame
df = pd.read_parquet(file_path)

# Restablecer el índice del DataFrame
df.reset_index(drop=True, inplace=True)

# Seleccionar solo las columnas deseadas( en este caso estan todas)
columns_to_keep = [
       'PersonaID','FechaNacimientoAAAAMM', 'Sexo', 'RegimenAfiliacionSGSSS',
       'Administradora', 'CodigoPertenenciaEtnica', 'municipioResidencia',
       'SemanasGestacion', 'EdadMadreMomentoDiagnostico',
       'FechaCertificadoAAAAMM', 'IpsDondeRealizaParto', 'Peso'
]


# Filtrar el DataFrame para mantener solo las columnas seleccionadas
df_filtered = df[columns_to_keep]

# Agregar una nueva columna con el recuento distintivo de 'personaid'
df_filtered['personaid_count'] = df_filtered.groupby('PersonaID')['PersonaID'].transform('count')

# Mostrar las primeras filas del DataFrame filtrado para verificar
#print(df_filtered.head())

# Mostrar los nombres de las columnas filtradas para verificar
print("Nombres de las columnas filtradas:")
print(df_filtered.columns)


# Mostrar las primeras filas del DataFrame con la nueva columna
#print(df_filtered.head())

# Agrupar por la columna 'personaid' y agregar la columna de recuento distintivo

# Agrupar por la columna 'personaid'
grouped_df = df_filtered.groupby('PersonaID').agg({
    'FechaNacimientoAAAAMM': 'first',
    'Sexo': 'first',
    'RegimenAfiliacionSGSSS': 'first',
    'Administradora': 'first',
    'CodigoPertenenciaEtnica': 'first',
    'municipioResidencia': 'first',
    'SemanasGestacion': 'first',
    'EdadMadreMomentoDiagnostico': 'first',
    'FechaCertificadoAAAAMM': 'first',
    'IpsDondeRealizaParto': 'first',
    'Peso': 'first',
    'personaid_count': 'first'

}).reset_index()

# Mostrar las primeras filas del DataFrame agrupado para verificar
#print(grouped_df.head().personaid_count)

# Guardar el DataFrame agrupado en un nuevo archivo Parquet
output_file_path = '/content/2.VITALES_agrupado0_2022.parquet'
grouped_df.to_parquet(output_file_path, engine='pyarrow')

print(f"El DataFrame agrupado ha sido guardado en {output_file_path}.")
```
